```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# ==============================================
# 1. AUTOENCODER BIASA
# ==============================================
class Autoencoder(nn.Module):
    def __init__(self, latent_dim=2):
        super(Autoencoder, self).__init__()

        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim)
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 28*28),
            nn.Sigmoid()  # Output antara 0-1
        )

    def forward(self, x):
        z = self.encoder(x)
        reconstruction = self.decoder(z)
        return reconstruction, z

# ==============================================
# 2. VARIATIONAL AUTOENCODER (VAE)
# ==============================================
class VAE(nn.Module):
    def __init__(self, latent_dim=2):
        super(VAE, self).__init__()

        # Encoder menghasilkan mean dan log variance
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU()
        )

        self.fc_mu = nn.Linear(256, latent_dim)  # Mean
        self.fc_logvar = nn.Linear(256, latent_dim)  # Log variance

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 28*28),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        """Reparameterization trick: z = mu + sigma * epsilon"""
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        z = mu + eps * std
        return z

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        reconstruction = self.decode(z)
        return reconstruction, mu, logvar

# ==============================================
# 3. LOSS FUNCTIONS
# ==============================================
def autoencoder_loss(reconstruction, original):
    """MSE loss untuk autoencoder biasa"""
    return F.mse_loss(reconstruction, original, reduction='sum')

def vae_loss(reconstruction, original, mu, logvar):
    """VAE loss = Reconstruction loss + KL divergence"""
    # Reconstruction loss (Binary Cross Entropy)
    recon_loss = F.binary_cross_entropy(reconstruction, original, reduction='sum')

    # KL divergence: -0.5 * sum(1 + log(sigma^2) - mu^2 - sigma^2)
    kl_divergence = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    return recon_loss + kl_divergence

# ==============================================
# 4. TRAINING FUNCTION
# ==============================================
def train_autoencoder(model, train_loader, optimizer, epoch, model_type='AE'):
    model.train()
    total_loss = 0

    for batch_idx, (data, _) in enumerate(tqdm(train_loader, desc=f'Epoch {epoch}')):
        data = data.view(-1, 28*28).to(device)
        optimizer.zero_grad()

        if model_type == 'VAE':
            reconstruction, mu, logvar = model(data)
            loss = vae_loss(reconstruction, data, mu, logvar)
        else:
            reconstruction, _ = model(data)
            loss = autoencoder_loss(reconstruction, data)

        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader.dataset)
    print(f'{model_type} Epoch {epoch}, Average Loss: {avg_loss:.4f}')
    return avg_loss

# ==============================================
# 5. VISUALISASI LATENT SPACE
# ==============================================
def visualize_latent_space(model, test_loader, model_type='VAE'):
    """Visualisasi distribusi latent space (hanya untuk latent_dim=2)"""
    model.eval()
    latents = []
    labels = []

    with torch.no_grad():
        for data, label in test_loader:
            data = data.view(-1, 28*28).to(device)

            if model_type == 'VAE':
                mu, _ = model.encode(data)
                latents.append(mu.cpu().numpy())
            else:
                _, z = model(data)
                latents.append(z.cpu().numpy())

            labels.append(label.numpy())

    latents = np.concatenate(latents, axis=0)
    labels = np.concatenate(labels, axis=0)

    # Plot
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(latents[:, 0], latents[:, 1], c=labels, cmap='tab10', alpha=0.6, s=5)
    plt.colorbar(scatter, label='Class')
    plt.xlabel('Latent Dimension 1')
    plt.ylabel('Latent Dimension 2')
    plt.title(f'{model_type} - Latent Space Distribution')
    plt.grid(True, alpha=0.3)

    # Label kelas Fashion-MNIST
    class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat',
                   'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

    plt.savefig(f'{model_type}_latent_space.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"Latent space plot saved as '{model_type}_latent_space.png'")

# ==============================================
# 6. LATENT SPACE ARITHMETIC (INTERPOLASI)
# ==============================================
def latent_space_interpolation(model, test_loader, model_type='VAE', n_steps=10):
    """Interpolasi antara 2 gambar di latent space"""
    model.eval()

    # Ambil 2 sampel berbeda
    data_iter = iter(test_loader)
    data1, label1 = next(data_iter)
    data2, label2 = next(data_iter)

    # Pilih 1 gambar dari masing-masing batch
    img1 = data1[0:1].view(-1, 28*28).to(device)
    img2 = data2[0:1].view(-1, 28*28).to(device)

    with torch.no_grad():
        if model_type == 'VAE':
            mu1, _ = model.encode(img1)
            mu2, _ = model.encode(img2)
            z1, z2 = mu1, mu2
        else:
            _, z1 = model(img1)
            _, z2 = model(img2)

    # Interpolasi
    fig, axes = plt.subplots(2, n_steps, figsize=(20, 4))

    for i, alpha in enumerate(np.linspace(0, 1, n_steps)):
        z_interpolated = (1 - alpha) * z1 + alpha * z2

        with torch.no_grad():
            if model_type == 'VAE':
                reconstruction = model.decode(z_interpolated)
            else:
                reconstruction = model.decoder(z_interpolated)

        img = reconstruction.cpu().view(28, 28).numpy()

        axes[0, i].imshow(img, cmap='gray')
        axes[0, i].axis('off')
        axes[0, i].set_title(f'α={alpha:.1f}', fontsize=8)

    # Tampilkan gambar asli di baris kedua
    axes[1, 0].imshow(img1.cpu().view(28, 28).numpy(), cmap='gray')
    axes[1, 0].set_title('Original 1', fontsize=8)
    axes[1, 0].axis('off')

    axes[1, -1].imshow(img2.cpu().view(28, 28).numpy(), cmap='gray')
    axes[1, -1].set_title('Original 2', fontsize=8)
    axes[1, -1].axis('off')

    for i in range(1, n_steps-1):
        axes[1, i].axis('off')

    plt.suptitle(f'{model_type} - Latent Space Interpolation')
    plt.tight_layout()
    plt.savefig(f'{model_type}_interpolation.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"Interpolation plot saved as '{model_type}_interpolation.png'")

# ==============================================
# 7. PERBANDINGAN REKONSTRUKSI
# ==============================================
def compare_reconstructions(ae_model, vae_model, test_loader, n_samples=10):
    """Bandingkan hasil rekonstruksi VAE vs Autoencoder"""
    ae_model.eval()
    vae_model.eval()

    data, _ = next(iter(test_loader))
    data = data[:n_samples].to(device)

    with torch.no_grad():
        # Autoencoder
        ae_recon, _ = ae_model(data.view(-1, 28*28))
        ae_recon = ae_recon.view(-1, 28, 28)

        # VAE
        vae_recon, _, _ = vae_model(data.view(-1, 28*28))
        vae_recon = vae_recon.view(-1, 28, 28)

    # Plot
    fig, axes = plt.subplots(3, n_samples, figsize=(20, 6))

    for i in range(n_samples):
        # Original
        axes[0, i].imshow(data[i].cpu().squeeze(), cmap='gray')
        axes[0, i].axis('off')
        if i == 0:
            axes[0, i].set_ylabel('Original', fontsize=12)

        # Autoencoder
        axes[1, i].imshow(ae_recon[i].cpu(), cmap='gray')
        axes[1, i].axis('off')
        if i == 0:
            axes[1, i].set_ylabel('Autoencoder', fontsize=12)

        # VAE
        axes[2, i].imshow(vae_recon[i].cpu(), cmap='gray')
        axes[2, i].axis('off')
        if i == 0:
            axes[2, i].set_ylabel('VAE', fontsize=12)

    plt.suptitle('Comparison: Autoencoder vs VAE Reconstruction')
    plt.tight_layout()
    plt.savefig('comparison_reconstruction.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("Comparison plot saved as 'comparison_reconstruction.png'")

# ==============================================
# 8. GENERATE GAMBAR BARU (HANYA VAE)
# ==============================================
def generate_new_images(vae_model, n_samples=20):
    """Generate gambar baru dengan sampling dari N(0,1)"""
    vae_model.eval()

    with torch.no_grad():
        # Sample dari distribusi normal
        z = torch.randn(n_samples, 2).to(device)
        generated = vae_model.decode(z)
        generated = generated.view(-1, 28, 28)

    # Plot
    fig, axes = plt.subplots(2, 10, figsize=(20, 4))

    for i in range(n_samples):
        row = i // 10
        col = i % 10
        axes[row, col].imshow(generated[i].cpu(), cmap='gray')
        axes[row, col].axis('off')

    plt.suptitle('VAE - Generated New Images (sampled from N(0,1))')
    plt.tight_layout()
    plt.savefig('vae_generated_images.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("Generated images saved as 'vae_generated_images.png'")

# ==============================================
# MAIN PROGRAM
# ==============================================
def main():
    # Hyperparameters
    BATCH_SIZE = 128
    LATENT_DIM = 2  # 2D untuk visualisasi
    LEARNING_RATE = 1e-3
    EPOCHS = 50

    # Load Fashion-MNIST dataset
    transform = transforms.ToTensor()

    train_dataset = datasets.FashionMNIST('./data', train=True, download=True, transform=transform)
    test_dataset = datasets.FashionMNIST('./data', train=False, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    print("Dataset loaded successfully!")
    print(f"Training samples: {len(train_dataset)}")
    print(f"Test samples: {len(test_dataset)}\n")

    # ==============================================
    # Train Autoencoder Biasa
    # ==============================================
    print("="*50)
    print("TRAINING AUTOENCODER")
    print("="*50)

    ae_model = Autoencoder(latent_dim=LATENT_DIM).to(device)
    ae_optimizer = optim.Adam(ae_model.parameters(), lr=LEARNING_RATE)

    for epoch in range(1, EPOCHS + 1):
        train_autoencoder(ae_model, train_loader, ae_optimizer, epoch, model_type='AE')

    print("\n")

    # ==============================================
    # Train VAE
    # ==============================================
    print("="*50)
    print("TRAINING VAE")
    print("="*50)

    vae_model = VAE(latent_dim=LATENT_DIM).to(device)
    vae_optimizer = optim.Adam(vae_model.parameters(), lr=LEARNING_RATE)

    for epoch in range(1, EPOCHS + 1):
        train_autoencoder(vae_model, train_loader, vae_optimizer, epoch, model_type='VAE')

    print("\n")

    # ==============================================
    # VISUALISASI DAN EVALUASI
    # ==============================================
    print("="*50)
    print("VISUALIZATION & EVALUATION")
    print("="*50)

    # 1. Visualisasi Latent Space
    print("\n1. Visualizing Latent Space Distribution...")
    visualize_latent_space(ae_model, test_loader, model_type='Autoencoder')
    visualize_latent_space(vae_model, test_loader, model_type='VAE')

    # 2. Latent Space Interpolation
    print("\n2. Performing Latent Space Interpolation...")
    latent_space_interpolation(ae_model, test_loader, model_type='Autoencoder', n_steps=10)
    latent_space_interpolation(vae_model, test_loader, model_type='VAE', n_steps=10)

    # 3. Perbandingan Rekonstruksi
    print("\n3. Comparing Reconstructions...")
    compare_reconstructions(ae_model, vae_model, test_loader, n_samples=10)

    # 4. Generate Gambar Baru (hanya VAE)
    print("\n4. Generating New Images with VAE...")
    generate_new_images(vae_model, n_samples=20)

    print("\n" + "="*50)
    print("ALL TASKS COMPLETED!")
    print("="*50)
    print("\nGenerated files:")
    print("- Autoencoder_latent_space.png")
    print("- VAE_latent_space.png")
    print("- Autoencoder_interpolation.png")
    print("- VAE_interpolation.png")
    print("- comparison_reconstruction.png")
    print("- vae_generated_images.png")

if __name__ == "__main__":
    main()
```
